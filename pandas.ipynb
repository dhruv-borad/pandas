# Pandas

```python
import pandas as pd
```

```python
df = pd.read_csv('US  E-commerce records 2020.csv')
df.head(5)
```

```python
# Read Headers
df.columns

# Read each Column
print(df[['City', 'Region', 'Sales']][0:5])

# Read Each Row
print(df.iloc[0:2])

df.loc[df['City'] == "Chicago"][0:5]

# Read a specific location (R,C)
print(df.iloc[1,7])

```

## Describe

```python
df.describe() # basic stats

# sort values by name 
df.sort_values(['City','Sales'],ascending=True)
```

## Modify data

```python
# you can also add columns and make new variable 
# below is just an exmaple. Does not make any sense.

df['New variable'] = df['Profit'] + df['Sales'] # add, subtract, revenue or any formula can be applied here.
df.head(5)

# drop new column
df = df.drop(columns='New variable')
df.head(2)

```

## Write to csv, excel, text

```python
# to csv
df.to_csv('write.csv',index=False)

# to excel
df.to_excel('write_xl.xlsx',index=False)

# to text
df.to_csv('write.txt',index=False,sep='\t')  # seperate by tab
```

## Filtering

```python
df.loc[df['City'] == 'Chicago'][0:2]
```

```python
df_filter = df.loc[(df['City'] == 'Chicago') & (df['Segment'] == 'Consumer') & (df['Sales'] > 100)][0:2]

# reset index values for filtered data
df_filter = df_filter.reset_index(drop=True)  # drop will remove previous index values.

# same step just using inplace instead of assigning to current df
df_filter.reset_index(drop=True,inplace=True)

df_filter
```

```python
# get all values containing this string
df.loc[df['City'].str.contains('Long')][0:5]

# get all values NOT containing this string
df.loc[~df['City'].str.contains('Long')][0:5]

import re
df.loc[df['City'].str.contains('Long|Los',regex=True)][0:5]  # containg Long or Los 

# with regex ignore case 
df.loc[df['City'].str.contains('long|los', flags=re.I, regex=True)][0:5]  # when bored to check case sensitiveness  
```

```python
# Conditional changes
df.loc[df['City'] == 'Chicago','City'] = 'Windy City'   # value replaced
df.loc[df['City'] == "Windy City"]
```

## Aggregate function

```python
df.groupby(['Segment']).mean().sort_values('Profit',ascending=False)

df.groupby(['Segment']).sum().sort_values('Quantity',ascending=False)

df['Count'] = 1
df.groupby(['City','Segment']).count()['Count'][0:15]
```

## Working with large dataset

```python
# take chunks of data
new_df = pd.DataFrame(columns=df.columns)
new_df

for df in pd.read_csv('US  E-commerce records 2020.csv',chunksize=5):
    new_df = pd.concat([df,new_df])
    print(new_df)
```
